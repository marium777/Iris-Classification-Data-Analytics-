```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/Unemployment in India.xlsx'
data = pd.read_excel(file_path)
```

```python
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
```

```python
cleaned_data = data.dropna()
```

```python
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=cleaned_data, marker='o')
```

```python
plt.axvline(pd.to_datetime('2020-03-24'), color='red', linestyle='--', label='Start of Major Lockdown')
plt.axvline(pd.to_datetime('2020-06-01'), color='blue', linestyle='--', label='End of Major Lockdown')
```

```python
plt.title('Unemployment Rate Trends by Region in India', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Unemployment Rate (%)', fontsize=14)
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.show()
```
